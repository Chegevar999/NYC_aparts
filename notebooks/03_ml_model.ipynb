# 03 ML-модель — KMeans кластеризация

```python
from pathlib import Path
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
df = pd.read_csv(ROOT / "data" / "processed" / "clean_real_estate_with_clusters.csv")
ml_cols = ["LotArea","OverallQual","OverallCond","YearBuilt","GrLivArea","FullBath","GarageCars","GarageArea","TotalBsmtSF","TotRmsAbvGrd"]
X = StandardScaler().fit_transform(df[ml_cols])
results = []
for k in range(2, 8):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X)
    results.append([k, model.inertia_, silhouette_score(X, labels)])
pd.DataFrame(results, columns=["k", "inertia", "silhouette"])
```

```python
df.groupby("Cluster")[ml_cols].mean().round(2)
```
